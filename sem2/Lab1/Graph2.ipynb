```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-10, 10, 0.01)
y=np.log((x**2+1)*(np.exp((-abs(x))/10)))/np.log(1+np.tan(1/(1+(np.sin(x))**2)))
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [0.99, 0.49, 0.35, 0.253, 0.18]
x1 = np.arange(1, 5, 0.01)
plt.errorbar(x, y, xerr=0.05, yerr=0.1)
plt.grid()
p1, v1=np.polyfit(x, y, deg=1, cov=True)
p2, v2=np.polyfit(x, y, deg=2, cov=True)
p1_f=np.poly1d(p1)
p2_f=np.poly1d(p2)
plt.plot(x, y, x1, p1_f(x1), x1, p2_f(x1))
plt.show()
```
