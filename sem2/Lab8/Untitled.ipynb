```python
import os
path='1'
abspath=os.path.abspath(path)
after_mas=[]
for name in os.listdir(abspath):
    full_name=os.path.join(abspath, name)
    type_file=os.path.splitext(full_name)
    after_mas.append(type_file[1])
    #print(after_mas)
for k in range (len(after_mas)):
    if after_mas[k]=='.txt':
        old_name=os.listdir(abspath)[k]
        new_name=os.path.splitext(old_name)[0]
        full_old=os.path.join(abspath, old_name)
        full_new=os.path.join(abspath, new_name)
        os.rename(full_old, full_new)
#print(os.listdir(abspath))
```

```python
import matplotlib.pyplot as plt
import pandas as pd
path='1'
try:
    os.mkdir('3')
except:
    pass
path3='3'
abspath3=os.path.abspath(path3)
for name in os.listdir(abspath):
    full_name=os.path.join(abspath, name)
    df=pd.read_csv(full_name, header=None)
    #print(df)
    df1=df[0]
    df2=pd.DataFrame()
    b=df.columns[1: ]
    a=len(df[0])
    for i in b:
        df2=pd.concat([df2, df[i]], axis=1)
    #print(df1, df2)
    plt.plot(df1, df2, 'o')
    new_name=os.path.splitext(name)[0]+'.jpeg'
    full_new_name=os.path.join(abspath3, new_name)
    #print(full_new_name)
    plt.savefig(full_new_name)
    plt.show()
    
```
