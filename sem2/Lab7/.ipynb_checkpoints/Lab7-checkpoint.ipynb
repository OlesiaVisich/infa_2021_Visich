```python
import pandas as pd
import matplotlib.pyplot as plt
students = pd.read_csv("StudentsPerformance.csv")
degrees=students["parental level of education"].unique()
data=[0]*len(degrees)
for i in range (len(degrees)):
    data[i]=students[students["parental level of education"]==degrees[i]]
for i, df in enumerate(data):
    plt.figure(figsize=(20,10))
    sp=plt.subplot(4, 3, i+1)
    sp.scatter(df["reading score"], df["math score"])
    #sp.legend(degrees[i])
    plt.title(degrees[i])
    plt.xlabel('Reading score')
    plt.ylabel('Math score')
for i, df in enumerate(data):
    plt.figure(figsize=(20,10))
    sp=plt.subplot(4, 3, i+7)
    sp.scatter(df["reading score"], df["writing score"])
    #sp.legend(degrees[i])
    plt.title(degrees[i])
    plt.xlabel('Reading score')
    plt.ylabel('Writing score')
plt.tight_layout()
plt.show()
```

```python
reading_data=[0]*len(degrees)
math_data=[0]*len(degrees)
writing_data=[0]*len(degrees)
for i, df in enumerate(data):
    reading_data[i]=df['reading score']
    math_data[i]=df['math score']
    writing_data[i]=df['writing score']
plt.figure(figsize=(30,10))
sp1=plt.subplot(1, 3, 1)
sp1.boxplot(reading_data, labels=degrees)
plt.title('reading score')
#print(len(writing_data), len(degrees))
sp2=plt.subplot(1, 3, 2)
sp2.boxplot(writing_data, labels=degrees)
plt.title('writing score')
sp3=plt.subplot(1, 3, 3)
sp3.boxplot(math_data, labels=degrees)
plt.title('math score')
plt.show()
```
