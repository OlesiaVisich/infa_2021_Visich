```python
df = pd.read_excel('task2_column_names.ods')
def rename (string):
    result=''
    if string[0]=='A':
        result+='контрольный образец, '
    elif string[0]=='S':
        result+='тестируемый образец, '
    result+='разбавлен в '
    result+=string[1:-1]
    result+=' раз(а), '
    if string[-1]=='+':
        result+='есть модулятор'
    elif string[-1]=='-':
        result+='нет модулятора'
    return result
namemas=[]
for i in range (1, len(df.columns)):
    namemas.append(rename(str(df.columns[i])))
print(namemas)
#rename('A3-')
```
